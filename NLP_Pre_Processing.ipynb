# 1. TOKENISATION

### Step 1: Prepare your workstation

```python
# Import all necessary libraries.
import nltk


# Install the required tokenisation function.
from nltk.tokenize import sent_tokenize
```

### Step 2: Tokenise the sentences

```python
# [1] Assign the raw text data to a variable.
text = """I went to many branches for help to find my child trust fund account however no one could find it until I visited the branch in Finchely road. Sangita Was amazing she was able to help me find my account. Amazing customer service and communication. She was very friendly and was able to help me solve my issue and answer all my queries. Thank you so much."""
```

```python

# [2] Tokenise the text data.
tokenized_sentence = sent_tokenize(text)

# [3] Check the result.
print(tokenized_sentence)
```

Notes: As you can see from the output above the text has been brokendown into sentences. In other words we have used sent_tokenize to tokensize the text into sentences which is stored as a list. However we feel it might be better suited to further break it down into a list of words.

```python
# Import the function.
from nltk.tokenize import word_tokenize

# Tokenise the text data.
tokenized_word = word_tokenize(text)

# Check the result.
print(tokenized_word)
```

Notes: we have further broken down the sentances into words using word_tokenize and created a list of words

### Step 3: Analyse individual words


The pre-processing allows us to analyse and compare the individual words. When working with a set of many reviews, this would allow you to make comparisons and calculations.

```python
# [1] Import the class.
from nltk.probability import FreqDist

# [2] Create a frequency distribution object.
freq_dist_of_words = FreqDist(tokenized_word)

# [3] Show the five most common elements in the data set.
freq_dist_of_words.most_common(5)
```

What we have done above is to count the most common words and the frequency of each. Notice that word_tokenize and most_common both recognise punction as words

```python
# [1] Import the package.
import matplotlib.pyplot as plt

# [2] Define the figure and axes.
fig, ax = plt.subplots(dpi=100)
fig.set_size_inches(18, 12)

# [3] Plot the data set.
freq_dist_of_words.plot(30, cumulative=False)

# [4] Set the labels for the axes
ax.set_xlabel('Counts', fontsize=20)
ax.set_ylabel('Samples', fontsize=20)

# [5] Display the result.
plt.show()
```

Above we have visualised the frequency of each word in the text using matplotlib

# 2. Normalisation 

Normalisation aims to change the textual data into a comparable and uniform format. For example, all characters should be lowercase, remove numbers or convert to textual representation, remove punctuation, strip white space, and remove general English stop words. Normalisation processes are based on a detailed and nuanced understanding of grammatical rules and norms. 

### 1. Remove stop words

```python
# [1] Import all necessary libraries.
import nltk

# [2] Download the stop words.
nltk.download('stopwords')

# [3] Import the package.
from nltk.corpus import stopwords

# [4] Create a set of English stop words.
stop_words=set(stopwords.words('english'))

# [5] Display the set.
print(stop_words)
```

```python

# [1] Create an empty list for the filtered words.
filtered_text = []

# [2] Create a tokenised word list.
tokenized_word = word_tokenize(text)

# [3] Filter the tokenised words.
for each_word in tokenized_word:
    if each_word not in stop_words:
        filtered_text.append(each_word)

# [4] Display the filtered list.
print('Tokenised list without stop words: {}'.format(filtered_text))
```

Notes: using stopwords we have now removed all of the stopwords from the text and are left with a list of all of the non stop words

### 2. Reduce words using stemming and lemmatisation

Under the umbrella of normalisation, there are two basic processes to reduce words to their root form: stemming and lemmatisation. Select the tabs to learn more about each reduction process.

between the two Lemmatisation is more accurate

### Using a lemmatiser

```python
# [1] Import the necessary class.
import nltk
from nltk.stem.wordnet import WordNetLemmatizer

# [2] Download the lemma corpus.
nltk.download('omw-1.4')

# [3] Create an instance of the class.
lemmatizer = WordNetLemmatizer()


# [4] Create an empty output list.
lemmatized_words_list = []
 

# [5] Apply lemmatisation to each tokenised word.
for each_word in filtered_text:
    lem_word = lemmatizer.lemmatize(each_word)
    lemmatized_words_list.append(lem_word)

# Display the output list.
print('Lemmatized Words list {}'.format(lemmatized_words_list))
```

Notes: Notice how all the plurals have been changed to singular (e.g. branches to branch'. However, unrecognised words (e.g. grandkids) are unchanged. A limitation of these pre-processing tools is that they are only as good as their underlying language corpora. 

# 3. Noise removal

Noise removal is a task-specific process rather than a language-specific process. It can:

 remove text file headers and footers


remove HTML, XML, markup, and metadata

 extract valuable data from other formats (e.g. JSON) or in databases.

Noise removal depends on how data is originally captured or wrapped. Pattern matching (i.e. checking a sequence of tokens for the presence of a pattern) can help speed up this process as it can take up to 80% of data preparation time. In a single block of text, we can strip the text of formatting by, for example, removing punctuation. 

Below I will use an example customer review to demonstrate how to use re to remove all punctuation

```python
# [1] Import the necessary module.
import re

# [2] Define some text. 
text = "Perfect! Buying a second. Using it to make a hot air balloon for an escape room adventure. Event in Oct. will share photos."


# [3] Filter out the specified punctuation.
no_punct = re.sub(r'[\.\?\!\,\:\;\"]', '', text)

# [4] Display the filtered text.
print(no_punct)
```

notes: as you can see from the above output all punctuation has been removed
